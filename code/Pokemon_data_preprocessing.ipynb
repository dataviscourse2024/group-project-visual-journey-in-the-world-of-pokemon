```python
import zipfile
import os

zip_file_path = 'pokemon_jpg.zip'

output_dir = 'pokemon_jpg'
with zipfile.ZipFile(zip_file_path, 'r') as z:
    z.extractall(output_dir)
```

```python
image_dir = 'pokemon_jpg/pokemon_jpg'

def map_image_filename(pokedex_number):
    return f"{pokedex_number}.jpg"

df['image_filename'] = df['pokedex_number'].apply(map_image_filename)

df.head()
```

```python
df['image_exists'] = df['image_filename'].apply(lambda x: os.path.exists(os.path.join(image_dir, x)))
```

```python
df_with_images = df[df['image_exists'] == True]
df_with_images.head()
```

```python
len(df_with_images)
```

```python
df.to_csv('pokemon_stats_with_images.csv', index=False)
```

## Pokemon Dataset with Team Combat

```python
df = pd.read_csv('combats.csv')
df.head()
```

```python
columns_to_read = ['pokedex_number','name','attack','defense','hp','sp_attack','sp_defense']
stats_df = pd.read_csv('pokemon_complete.csv',usecols=columns_to_read)
comeback_df = pd.read_csv('combats.csv')

comeback_df = pd.merge(comeback_df, stats_df, left_on='First_pokemon', right_on='pokedex_number', suffixes=('', '_first'))

comeback_df = pd.merge(comeback_df, stats_df, left_on='Second_pokemon', right_on='pokedex_number', suffixes=('_first', '_second'))

comeback_df = comeback_df.drop(columns=['pokedex_number_first', 'pokedex_number_second'])

comeback_df.head()
```

```python
comeback_df = pd.merge(comeback_df, stats_df[['pokedex_number', 'name']], left_on='Winner', right_on='pokedex_number', suffixes=('', '_winner'))

comeback_df = comeback_df.drop(columns=['First_pokemon','Second_pokemon','Winner'])

comeback_df = comeback_df.rename(columns={'name': 'winner_name'})

comeback_df.head()
```

```python
len(comeback_df)
```

```python
comeback_df.to_csv('comeback_results.csv', index=False)
```
